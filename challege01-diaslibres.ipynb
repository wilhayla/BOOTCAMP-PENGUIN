## Challege ##
¿Que quiero que el programa haga?
1. Quiero que me muestre un menu donde tenga 4 opciones. 
- Agregar tarea nueva
- Modificar el estado de la tarea
- Eliminar una tarea
- mostrar todas las tareas.

2. Un mensaje donde me pide elegir la opcion del menu.
3. Si elijo la opcion 1 (agregar tarea), tipeo por teclado una tarea nueva. Y luego pedir por teclado si esa tarea esta en pendiente, en progreso o completado.
4. Si elijo la opcion 2 (modificar el estado de la tarea), el programa me pregunta que tarea deseo modificar y lo tipeo por teclado. Luego me salta un mensaje donde me pregunta que opcion de estado quisiera modificar, tipero pendiente, en progreso o completado.
5. Si elijo la opcion 3 (eliminar tarea). Me salta un mensaje (¿Que tarea quieres eliminar?), tipeo la tarea a eliminar.
6. Si elijo la opcion 6 (mostrar todas las tareas), el programa me muestra en pantalla todas las tareas.

- para almacenar estas tareas tengo que crear un diccionario.
- para elegir las opciones del menu tengo que crear un bucle.

## COMO ESTRUCTURAR EL PROGRAMA

1. Lo voy a estructurar en una clase.
2. Establezco el diccionario como atributo de la clase.



```python
class GestionTareas:
    tareas = {}
    
    def menu(self):
        print('1. Agregar una tarea nueva.')
        print('2. Modificar el estado de una tarea.')
        print('3. Eliminar una tarea.')
        print('4. Mostrar todas las tareas y sus estados.')
        print('5. Salir.')
                
    def agregar_tareas(self):
        while True:
            salir = input('presione "s" para salir o cualquier tecla para continuar: ')
            if salir == 's':
                break
            self.tarea = input('Agrega una tarea: ')
            self.estado = input('Estado de la tarea: ')
            GestionTareas.tareas[self.tarea] = self.estado
            print(GestionTareas.tareas)
    
    def modificar_estado(self):
        tarea_modificar = input('Que tareas deseas modificar?: ')
        if tarea_modificar in GestionTareas.tareas:
            estado_modificar = input(f'Modifica el estado de la tarea {tarea_modificar}')
            GestionTareas.tareas[tarea_modificar] = estado_modificar
            print(GestionTareas.tareas)
            print('Estado actualizado')
        else:
             print('No existe esta tarea en tu agenda.')
    
    def eliminar_tarea(self):
        tarea_eliminar = input('Que tarea deseas eliminar?: ')
        if tarea_eliminar in GestionTareas.tareas:
            del GestionTareas.tareas[tarea_eliminar]
        else:
            print('No existe esta tarea en tu agenda.')
        print(GestionTareas.tareas)
    
    def mostrar_tareas(self):
        print('Lista de las tareas.')
        print(GestionTareas.tareas)
        
            
# MAIN                                   
gestion1 = GestionTareas()
gestion1.menu()
        
while True:
    opcion = int(input('Elija una opcion: '))
    if opcion == 1:
        gestion1.agregar_tareas()
    elif opcion == 2:
        gestion1.modificar_estado()
    elif opcion == 3:
        gestion1.eliminar_tarea()
    elif opcion == 4:
         gestion1.mostrar_tareas()
    elif opcion == 5:
        break 
```
